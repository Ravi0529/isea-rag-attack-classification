```python
import pandas as pd

s = pd.read_parquet("../data/processed/sessions.parquet")
print("sessions:", len(s))
print(
    s.sort_values("event_count", ascending=False)
    .head(10)[
        [
            "session_id",
            "src_ip",
            "tool",
            "event_count",
            "duration_s",
            "rps",
            "indicator_hits",
        ]
    ]
    .to_string(index=False)
)
```
